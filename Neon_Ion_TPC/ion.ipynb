```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
VEL = np.loadtxt("Neon_Ion_Vel_Hornbeck.csv",delimiter=',')
DIF = np.loadtxt("Neon_Ion_Diffusion_skillerud_Larson.csv",delimiter=',',skiprows=1)
```

```python
plt.figure(figsize=(7,7))
plt.title(r'Ne$^+$ velocity in neon',fontsize=20)
plt.scatter(VEL[:,0]*760,VEL[:,1]*1e4,label='Hornbeck \n measurement',color='darkblue')

Ke = DIF[:,7]*1e-3
vely = np.sqrt((Ke*2*1.6022e-19)/3.35e-26)
redu = (DIF[:,0]*2.6867e25*1e-6*1e-17)/760

plt.scatter(redu,vely,label='Skullerud & Larsen \n calculation',color='darkred')
#plt.plot(velx,vely,color='g',label='calculation')
plt.xlabel('Reduced Field [V/cm/bar]',fontsize=20)
plt.xticks(fontsize=15)

plt.ylabel('Velocity [cm/s]',fontsize=20)
plt.yticks(fontsize=15)

plt.loglog()
plt.xlim(1e0,0.9e6)
plt.ylim(4e2,0.9e6)
plt.legend(loc='upper left', fontsize=15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.title(r'Ne$^+$ Diffusion calculation Skullerud & Larsen',fontsize=16)

redu = (DIF[:,0]*2.6867e25*1e-6*1e-17)/760
Dl = (DIF[:,3]*1e-3*DIF[:,1]*3.773e-5)*1e4
plt.scatter(redu,Dl,label="Longitudinal")

Dl = (DIF[:,5]*1e-3*DIF[:,1]*3.773e-5)*1e4
plt.scatter(redu,Dl,label="Transverse")

#plt.plot(velx,vely,color='g',label='calculation')
plt.xlabel('Reduced Field [V/cm/bar]',fontsize=20)
plt.xticks(fontsize=15)

plt.ylabel(r'Diffusion [cm$^2$/s]',fontsize=20)
plt.yticks(fontsize=15)

#plt.loglog()
#plt.xlim(1e0,0.9e6)
#plt.ylim(4e2,0.9e6)
plt.legend(loc='upper left', fontsize=15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.title(r'Ne$^+$ Diffusion calculation Skullerud & Larsen',fontsize=16)
Ke = DIF[:,7]*1e-3
vely = np.sqrt((Ke*2*1.6022e-19)/3.35e-26)

redu = (DIF[:,0]*2.6867e25*1e-6*1e-17)/760
Dl = (DIF[:,3]*1e-3*DIF[:,1]*3.773e-5)*1e4

dlstar = np.sqrt(2*Dl/vely)
plt.scatter(redu,dlstar,label="Longitudinal")

Dl = (DIF[:,5]*1e-3*DIF[:,1]*3.773e-5)*1e4
dlstar = np.sqrt(2*Dl/vely)
plt.scatter(redu,dlstar,label="Transverse")

#plt.plot(velx,vely,color='g',label='calculation')
plt.xlabel('Reduced Field [V/cm/bar]',fontsize=20)
plt.xticks(fontsize=15)

plt.ylabel(r'Reduced Diffusion $\sqrt{cm}$',fontsize=20)
plt.yticks(fontsize=15)

#plt.loglog()
#plt.xlim(1e0,0.9e6)
plt.ylim(0.007,0.022)
plt.legend(loc='upper left', fontsize=15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.title(r'Ne$^+$ Energy calculation Skullerud & Larsen',fontsize=16)
Ke = DIF[:,7]*1e-3

redu = (DIF[:,0]*2.6867e25*1e-6*1e-17)/760

plt.scatter(redu,Ke,color='darkgreen')
plt.axhline(0.025851,color='r')
plt.text(10,0.03,"Thermal limit",color='r',fontsize=15)

#plt.plot(velx,vely,color='g',label='calculation')
plt.xlabel('Reduced Field [V/cm/bar]',fontsize=20)
plt.xticks(fontsize=15)

plt.ylabel(r'Ion Energy [eV]',fontsize=20)
plt.yticks(fontsize=15)

#plt.semilogy()
plt.loglog()
#plt.xlim(1e0,0.9e6)
#plt.ylim(0.007,0.022)
#plt.legend(loc='upper left', fontsize=15)
plt.grid()
plt.show()
```

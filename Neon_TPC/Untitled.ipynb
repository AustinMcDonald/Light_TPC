```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class MERT_Props():
    def __init__(self):
        Efield   = np.array([])
        W        = np.array([])
        W_er     = np.array([])
        Dl       = np.array([])
        Dl_er    = np.array([])
        Dt       = np.array([])
        Dt_er    = np.array([])
        Dls      = np.array([])
        Dls_er   = np.array([])
        Dts      = np.array([])
        Dts_er   = np.array([])
        MCT      = np.array([])
```

```python
data = np.load('Neon_TPC.npy').item()
```

```python
data.keys()
```

```python
omega = np.array([0, 1.525e6, 3.812e6, 7.624e6, 1.144e7, 1.525e7]) # in Hz
```

```python
np.mean(data['0.0/'].MCT*omega[0])
```

```python
np.mean(data['1.0/'].MCT*omega[1])
```

```python
np.mean(data['10.0/'].MCT*omega[5])
```

```python
plt.figure(figsize=(7,7))
ti=1e6
plt.scatter(data['1.0/'].Efield,(data['1.0/'].MCT/ti*omega[1])**2,label="0.1T")
plt.scatter(data['2.5/'].Efield,(data['2.5/'].MCT/ti*omega[2])**2,label="0.25T")
plt.scatter(data['5.0/'].Efield,(data['5.0/'].MCT/ti*omega[3])**2,label="0.5T")
plt.scatter(data['7.5/'].Efield,(data['7.5/'].MCT/ti*omega[4])**2,label="0.75T")
plt.scatter(data['10.0/'].Efield,(data['10.0/'].MCT/ti*omega[5])**2,label="1.0T")
plt.semilogy()

plt.legend(loc='upper right',fontsize=15,ncol=2)

plt.tick_params(axis="x", labelsize=20)
plt.tick_params(axis="y", labelsize=20)
plt.xlabel(r'Reduced Drift Field [V cm$^{-1}$ bar$^{-1}$]',fontsize=20)
plt.ylabel(r'omega tau',fontsize=20)

plt.show()
```

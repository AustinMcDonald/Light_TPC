```python
def BOLTZ_EXTRACT(data):
    off = -5
    E  = data[:,5+off]
    Vz = data[:,6+off]
    Vzerr = Vz*data[:,7+off]/100
    Dl = data[:,8+off]
    Dlerr = Vz*data[:,9+off]/100
    Dt = data[:,10+off]
    Dterr = Vz*data[:,11+off]/100
    Dltpc = data[:,12+off]
    Dltpcerr = Vz*data[:,13+off]/100
    Dttpc = data[:,14+off]
    Dttpcerr = Vz*data[:,15+off]/100
    
    drop = np.where(Vz == 0)[0]
    E        = np.delete(E,drop)
    Vz       = np.delete(Vz,drop)
    Vzerr    = np.delete(Vzerr,drop)
    Dl       = np.delete(Dl,drop)
    Dlerr    = np.delete(Dlerr,drop)
    Dt       = np.delete(Dt,drop)
    Dterr    = np.delete(Dterr,drop)
    Dltpc    = np.delete(Dltpc,drop)
    Dltpcerr = np.delete(Dltpcerr,drop)
    Dttpc    = np.delete(Dttpc,drop)
    Dttpcerr = np.delete(Dttpcerr,drop)

    SORT = E.argsort()
    E        = E[SORT]
    Vz       = Vz[SORT]
    Vzerr    = Vzerr[SORT]
    Dl       = Dl[SORT]
    Dlerr    = Dlerr[SORT]
    Dt       = Dt[SORT]
    Dterr    = Dterr[SORT]
    Dltpc    = Dltpc[SORT]
    Dltpperr = Dltpcerr[SORT]
    Dttpc    = Dttpc[SORT]
    Dttpcerr = Dttpcerr[SORT]

    return E, Vz, Vzerr, Dl, Dlerr, Dt, Dterr, Dltpc, Dltpcerr, Dttpc, Dttpcerr

def MERT(epsilon, A, D, F, A1):
    alpha  = 27.292
    k = np.sqrt((epsilon)/(13.605))
    
    eta0 = -A*k*(1+(4*alpha)/(3)*k**2*np.log(k)) \
            -(np.pi*alpha)/(3)*k**2 + D*k**3 + F*k**4
    
    eta1 = (np.pi)/(15)*alpha*k**2 - A1*k**3
        
    Qm = (4*np.pi)/(k**2)*(np.sin(np.arctan(eta0)-np.arctan(eta1)))**2
    
    Qt = (4*np.pi)/(k**2)*(np.sin(np.arctan(eta0)))**2
    
    return Qm*(5.29e-11)**2*1e20, Qt*(5.29e-11)**2*1e20

def WEIGHT_Q(eV, Qm, BashBoltzQm, Lamda, eV0):
    WeightQm  = (1-np.tanh(Lamda*(eV-eV0)))/2
    WeightBB  = (1+np.tanh(Lamda*(eV-eV0)))/2
    
    NewBashQm = BashBoltzQm*WeightBB
    NewMERTQm = Qm*WeightQm
    NewQm = NewBashQm+NewMERTQm
    return NewQm
    
def HYBRID_X_SECTIONS(MB_EMTx, MB_EMTy, MB_ETx, MB_ETy, A,D,F,A1,Lambda,eV0):
    Qm_MERT, Qt_MERT = MERT(MB_EMTx, A,D,F,A1)
    New_Qm = WEIGHT_Q(MB_EMTx, Qm_MERT, MB_EMTy, Lambda, eV0)
    Qm_MERT, Qt_MERT = MERT(MB_ETx, A,D,F,A1)
    New_Qt = WEIGHT_Q(MB_ETx, Qt_MERT, MB_ETy, Lambda, eV0)
    
    return MB_EMTx, New_Qm, MB_ETx, New_Qt
```
